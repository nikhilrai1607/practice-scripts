```python
cities, years = [],[]

for game in open('games.txt','r'):
    words = game.split()
    
    city = ' '.join(words[:-1])
    year = words[-1].strip('()') 
    
    cities.append(city)
    years.append(year)
```

```python
geolocator = geopy.geocoders.Nominatim()
locations = {}

for city in cities:
    print('Locating ',city)
    locations[city] = geolocator.geocode(city.split('/'))
```

```python
pp.figure(figsize=(10,5))

world = Basemap()
world.drawcoastlines(linewidth=0.25)
world.drawcoastlines(linewidth=0.25)
```

```python
pp.figure(figsize=(10,5))

world = Basemap()
world.drawcoastlines(linewidth=0.25)
world.drawcoastlines(linewidth=0.25)

for city,pos in locations.items():
    world.plot(pos.longitude,pos.latitude,'r.',markersize=7,latlon=True)
```

```python
locations['Athens'] = geolocator.geocode('Athens, Greece')
locations['Melbourne / Stockholm'] = geolocator.geocode('Melbourne, Australia')
locations['Rome'] = geolocator.geocode('Rome, Italy')
```

```python
locations
```

```python
pp.figure(figsize=(10,5))

world = Basemap()
world.drawcoastlines(linewidth=0.25)
world.drawcoastlines(linewidth=0.25)

for city,pos in locations.items():
    world.plot(pos.longitude,pos.latitude,'r.',markersize=7,latlon=True)
```

```python
#Enumeration of list
for i,city in enumerate(cities[:10]):
    print(i,city)
```

```python
# Sorted list
for city in sorted(cities[:10]):
    print(city)
```

```python
# Sort based on the a criteria
for city in sorted(cities[:10],key=len):
    print(city)
```

```python
#reverse enumeration
for i,city in enumerate(reversed(cities[:10])):
    print(i,city)
```

```python
#zipping 2 lists together
for year,city in zip(years[:10],cities[:10]):
    print(year,city)
```
